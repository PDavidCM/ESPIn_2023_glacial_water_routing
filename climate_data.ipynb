```python
import numpy as np
from matplotlib import pyplot as plt
import xarray as xr
```

```python
ds = xr.open_dataset('data.nc')
```

```python
ds
```

```python
fig,ax=plt.subplots()
ds.smlt.mean(dim=['latitude','longitude']).plot(ax=ax,label='melt production')
plt.legend()
```

```python
fig,ax=plt.subplots()
ds.t2m.mean(dim=['latitude','longitude']).plot(ax=ax,label='temperature')
plt.legend()
```

```python
fig,ax=plt.subplots()
ds.smlt.mean(dim='time').plot(ax=ax,cbar_kwargs={'label':'melt produced in 2022 (mm w.e.)'})
```
